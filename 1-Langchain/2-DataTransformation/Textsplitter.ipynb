```python
## pdf file

from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('attention.pdf')
docs=loader.load()
docs
```

```python
## split text by characters

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=50)
final=text_splitter.split_documents(docs)
final
```

```python
## text loader

from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n",chunk_size=20,chunk_overlap=5)
text_splitter.split_documents(docs)
```

```python
speech=""

with open("speech.txt") as f:
    speech=f.read()

text_splitter=CharacterTextSplitter(chunk_size=25,chunk_overlap=5)
text=text_splitter.create_documents([speech])
print(text[0])
print(text[1])
```

```python
from langchain_text_splitters import HTMLHeaderTextSplitter 

html_str="""
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Sample HTML Document</title>
</head>
<body>
  <h1>Welcome to LangChain Data Ingestion</h1>

  <p>This is a sample HTML file for testing document loaders.</p>

  <h2>Technologies</h2>
  <ul>
    <li>Python</li>
    <li>LangChain</li>
    <li>Gemini</li>
    <li>Vector Database</li>
  </ul>

  <h2>Description</h2>
  <p>
    LangChain enables developers to build applications using large language
    models. It supports document loading, text splitting, embeddings,
    vector stores, and retrieval augmented generation (RAG).
  </p>

  <footer>
    <p>Created for Data Ingestion Practice</p>
  </footer>
</body>
</html>
"""

headers_to_split_on=[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_header_split=html_splitter.split_text(html_str)
html_header_split
```

```python

import json
import requests
json_data=requests.get("https://api.smith.langchain.com/openapi.json").json()
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter=RecursiveJsonSplitter(max_chunk_size=300)
json_chunks=json_splitter.split_json(json_data)
```
