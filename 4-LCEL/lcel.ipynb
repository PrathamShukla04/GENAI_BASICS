```python
import os
from dotenv import load_dotenv
load_dotenv()


gemini_api_key = os.getenv("GEMINI_API_KEY")


groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="llama-3.3-70b-versatile",groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage,SystemMessage

messages=[
    SystemMessage(content="Translate following into French"),
    HumanMessage(content="Hello ,How are you")
]
response=model.invoke(messages)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
parser.invoke(response)
```

```python
## lcel used to chain components

chain=model|parser
chain.invoke(messages)
```

```python
## prompt templates

from langchain_core.prompts import ChatPromptTemplate

generic_temp="Translate the following into {language}:"

prompt=ChatPromptTemplate.from_messages(
    [("system",generic_temp),("user","{text}")]
)
```

```python
result=prompt.invoke({"language":"Bengali", "text":"Hello"})
result.to_messages()
```

```python
chain=prompt|model|parser
```

```python
chain.invoke({"language":"Bengali", "text":"Hello"})
```
